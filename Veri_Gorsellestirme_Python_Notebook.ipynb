# Veri Görselleştirme Projesi - Python ile Görselleştirme



Kullanılan grafik türleri:

1. Box Plot
2. Violin Plot
3. Line Chart
4. Bar Chart
5. Pie Chart

Kullanılan Python kütüphaneleri:

- Matplotlib
- Pyplot
- Seaborn
- Plotly

Veri seti olarak `tips` veri seti kullanılmıştır. Bu veri seti restoran müşterilerinin toplam hesap, bahşiş, cinsiyet, sigara kullanımı, gün, öğün zamanı ve kişi sayısı bilgilerini içerir.

```python
# Gerekli kütüphanelerin içe aktarılması
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Grafiklerin daha düzenli görünmesi için tema ayarı
sns.set_theme(style="whitegrid")

# Veri setini yükleme
df = px.data.tips()

# İlk 5 satırı görüntüleme
df.head()
```

## 1. Veri Setini Tanıma

Bu bölümde veri setinin genel yapısı incelenmiştir. Veri setindeki sütunlar, veri tipleri ve temel istatistiksel bilgiler görüntülenmiştir.

```python
# Veri setinin boyutu
print("Satır ve sütun sayısı:", df.shape)

# Sütun bilgileri
df.info()
```

```python
# Sayısal değişkenlerin temel istatistikleri
df.describe()
```

```python
# Kategorik değişkenlerin değer sayıları
for col in ["sex", "smoker", "day", "time"]:
    print(f"\n{col} sütunu değerleri:")
    print(df[col].value_counts())
```

## 2. Box Plot

Box plot, sayısal bir değişkenin dağılımını ve aykırı değerlerini incelemek için kullanılır. Bu örnekte haftanın günlerine göre toplam hesap tutarlarının dağılımı incelenmiştir.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="day", y="total_bill")
plt.title("Günlere Göre Toplam Hesap Dağılımı - Box Plot")
plt.xlabel("Gün")
plt.ylabel("Toplam Hesap")
plt.show()
```

**Yorum:** Bu grafik, farklı günlerdeki toplam hesap tutarlarının dağılımını gösterir. Kutuların genişliği ve aykırı noktalar, hangi günlerde hesap tutarlarının daha değişken olduğunu anlamaya yardımcı olur.

## 3. Violin Plot

Violin plot, box plot'a benzer ancak verinin yoğunluk dağılımını da gösterir. Bu örnekte cinsiyete göre toplam hesap tutarlarının dağılımı incelenmiştir.

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x="sex", y="total_bill")
plt.title("Cinsiyete Göre Toplam Hesap Dağılımı - Violin Plot")
plt.xlabel("Cinsiyet")
plt.ylabel("Toplam Hesap")
plt.show()
```

**Yorum:** Grafikte geniş bölgeler, o değer aralığında daha fazla veri olduğunu gösterir. Violin plot sayesinde sadece ortalama veya medyan değil, dağılımın şekli de görülebilir.

## 4. Line Chart

Line chart, verinin belirli bir sıra veya zaman içindeki değişimini göstermek için kullanılır. Bu örnekte kayıt sırasına göre toplam hesap tutarındaki değişim gösterilmiştir.

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["total_bill"], marker="o", linewidth=1, markersize=3)
plt.title("Kayıt Sırasına Göre Toplam Hesap Değişimi - Line Chart")
plt.xlabel("Kayıt Sırası")
plt.ylabel("Toplam Hesap")
plt.show()
```

**Yorum:** Bu grafik, toplam hesap değerlerinin kayıt sırasına göre nasıl değiştiğini gösterir. Zaman değişkeni olmayan veri setlerinde indeks sırası üzerinden genel değişim izlenebilir.

## 5. Bar Chart

Bar chart, kategoriler arasındaki karşılaştırmaları göstermek için kullanılır. Bu örnekte haftanın günlerine göre ortalama bahşiş miktarı karşılaştırılmıştır.

```python
avg_tip_by_day = df.groupby("day", as_index=False)["tip"].mean()

plt.figure(figsize=(8, 5))
sns.barplot(data=avg_tip_by_day, x="day", y="tip")
plt.title("Günlere Göre Ortalama Bahşiş Miktarı - Bar Chart")
plt.xlabel("Gün")
plt.ylabel("Ortalama Bahşiş")
plt.show()

avg_tip_by_day
```

**Yorum:** Bu grafik, hangi günlerde ortalama bahşiş miktarının daha yüksek olduğunu karşılaştırmayı sağlar. Kategorik karşılaştırmalar için bar chart oldukça uygundur.

## 6. Pie Chart

Pie chart, kategorilerin toplam içindeki oranlarını göstermek için kullanılır. Bu örnekte sigara içen ve içmeyen müşterilerin oranı gösterilmiştir.

```python
smoker_counts = df["smoker"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(smoker_counts, labels=smoker_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Sigara İçen ve İçmeyen Müşteri Oranı - Pie Chart")
plt.show()

smoker_counts
```

**Yorum:** Pasta grafiği, sigara içen ve içmeyen müşterilerin toplam müşteri sayısı içindeki oranını gösterir. Az sayıda kategori olduğu için pie chart bu örnek için uygundur.

## 7. Plotly ile Etkileşimli Grafik Örneği

Plotly, etkileşimli grafikler oluşturmak için kullanılır. Aşağıdaki örnekte günlere göre toplam hesap dağılımı interaktif box plot ile gösterilmiştir.

```python
import warnings
warnings.filterwarnings("ignore")
fig = px.box(
    df,
    x="day",
    y="total_bill",
    color="day",
    title="Günlere Göre Toplam Hesap Dağılımı - Plotly Box Plot"
)
fig.show()
```

## 8. Genel Sonuç

Bu çalışmada `tips` veri seti kullanılarak beş temel grafik türü Python ortamında uygulanmıştır. Box plot ile günlere göre hesap dağılımı ve aykırı değerler incelenmiştir. Violin plot ile cinsiyete göre hesap tutarlarının yoğunluk dağılımı analiz edilmiştir. Line chart ile toplam hesapların kayıt sırasına göre değişimi gösterilmiştir. Bar chart ile günlere göre ortalama bahşiş miktarı karşılaştırılmıştır. Pie chart ile sigara içen ve içmeyen müşterilerin oranı görselleştirilmiştir.

Bu grafikler sayesinde farklı veri tipleri için hangi görselleştirme yöntemlerinin daha uygun olduğu uygulamalı olarak gösterilmiştir.
